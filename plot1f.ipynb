```python
from matplotlib import pyplot as plt
import numpy as np 
from scipy.stats import norm
import matplotlib.mlab as mlab
from scipy.optimize import leastsq
import math
```

```python
%matplotlib inline
a = np.genfromtxt('plot_avg10000.dat')
plt.title("PDF of distribution of sum,dx=1")
plt.hist(a*10000,bins=np.arange(4850,5150,1),density='True',histtype='stepfilled')
plt.xlabel("Sum of 10000 random numbers")
plt.show()
```

```python
%matplotlib inline
plt.title("CDF of distribution of mean,dx=1")
plt.hist(a*10000,bins=np.arange(4850,5150,1),histtype='step',density='True',cumulative='True')
plt.show()
```

```python
%matplotlib inline

a = np.genfromtxt('plot_avg_discrete10000.dat')
plt.title("PDF of sum of discrete random walk,dx=1")
mean=np.mean(a*10000)
stddev=np.std(a*10000)
x=np.arange(-449,450,2)
plt.hist(a*10000,bins=np.arange(-449,450,1),histtype='stepfilled',density="True",stacked='true')
plt.plot(x, np.exp(-0.5*((x-mean)/stddev)**2)/(stddev*math.sqrt(2*math.pi)), color='black',label='Fitted Gaussian, mean={} stddev={}'.format(mean,stddev))
plt.legend()
plt.xlabel("Position after 10000 steps" )
plt.show()
```
